```python
#from skimage.morphology import erosion, dilation, opening, closing, white_tophat
#from skimage.morphology import black_tophat, skeletonize, convex_hull_image
#from skimage.morphology import disk
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Loaded a set of images
root_dir = "training/"
image_dir = root_dir + "images/"
```

```python
import os
files = os.listdir(image_dir)
```

```python
def load_image(infilename):
    data = cv2.imread(infilename)
    return data
```

```python
imgs = [load_image(image_dir + files[i]) for i in range(len(files))]
```

```python
kernel = np.ones((7,7),np.float32)/49
dst = cv2.filter2D(imgs[3],-1,kernel)
```

```python
import matplotlib.pyplot as plt
import scipy.misc

for i in range(len(files)):
    dst = cv2.filter2D(imgs[i],-1,kernel)
    scipy.misc.imsave('smoothed/'+'im'+str(i)+'.png', dst)
```

```python
fig1 = plt.figure(figsize=(10, 10))
plt.imshow(dst, cmap='Greys_r')
plt.show()
```
